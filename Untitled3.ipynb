```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
import pandas as pd

# Read the CDV file into a pandas DataFrame
df = pd.read_csv('magic04.data', sep=',')

# Print the DataFrame
print(df)
print(df.columns)

# filtered_df = df['10'].replace({'g': -1, 'h': 1})
```

```python
# Replace 'g' with -1 and 'h' with 1 in the 'g' column
df['g'].replace({'g': -1, 'h': 1}, inplace=True)

# Print the modified DataFrame
print(df)
```

```python
print(len(df))
```

```python
import matplotlib.pyplot as plt

# Assuming df is your DataFrame containing the data
plt.scatter(df['28.7967'], df['16.0021'], c=df['g'], cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Scatter Plot of Features 1 and 2')
plt.colorbar(label='Label (-1 for g, 1 for h)')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.drop('g',axis=1), df['g'], test_size=0.5, random_state=101)

print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
print(X_train)
print(y_train)
#print(X_train[:0])
```

```python
def PercentCorrect(Inputs, targets, weights):
  N = len(targets)
  nCorrect = 0
  for n in range(N):
    OneInput = Inputs[n,:]
    if (targets[n] * np.dot(OneInput, weights) > 0):
      nCorrect +=1
  return 100*nCorrect/N

# Example usage
# Assuming weights are already defined
np.random.seed(123)
# w = np.random.randn(10)  # Randomly initialized weights
w = np.ones(10)
print('Initial Percentage Correct:', PercentCorrect(X_train.to_numpy(), y_train.to_numpy(), w))

MaxIter = 2000
alpha = 0.0001
Ntrain = len(X_train)  # Number of samples in the training set

P_train = np.zeros(MaxIter)
P_test = np.zeros(MaxIter)

for iter in range(MaxIter):
    r =  np.floor(np.random.rand()*Ntrain).astype(int)# Randomly select an index
    x = X_train.iloc[r, :]  # Select a single row

    if (y_train.iloc[r] * np.dot(x, w) < 0):
        w += alpha * y_train.iloc[r] * x
    P_train[iter] = PercentCorrect(X_train.to_numpy(), y_train.to_numpy(), w)
    P_test[iter] = PercentCorrect(X_test.to_numpy(), y_test.to_numpy(), w)
    print("iter: ", iter , " , Train_",P_train[iter], " , Test_",P_test[iter])

print('Percentage Correct After Training on Training Set:', PercentCorrect(X_train.to_numpy(), y_train.to_numpy(), w))
print('Percentage Correct After Training on Test Set:', PercentCorrect(X_test.to_numpy(), y_test.to_numpy(), w))
```

```python
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(range(MaxIter), P_train, 'b', label = "Training")
ax.plot(range(MaxIter), P_test, 'r', label = "Test")
ax.grid(True)
ax.legend()
ax.set_title('Perceptron Learning')
ax.set_ylabel('Training and Test Accuracies', fontsize=14)
ax.set_xlabel('Iteration', fontsize=14)
plt.savefig('learningCurves.png')
```
